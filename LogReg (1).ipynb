# Logistic Regression Project
## Introduction
This project explores how logistic regression can be used to predict whether an individual has insurance based on their age. The dataset used is simple and synthetic, where individuals above the age of 40 are more likely to have insurance. This analysis demonstrates the fundamentals of classification using logistic regression, including data preparation, model training, and evaluation.

```python
import pandas as pd 
import numpy as np
```

```python
insurance = pd.read_csv('age_insurance_dataset.csv')
insurance 
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(x='age', y='insurance', data=insurance, color='red')
plt.title('AGE VS INSURANCE')
plt.xlabel('age')
plt.ylabel('insurance')
```

```python
X = insurance[['age']] 
y = insurance['insurance']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
ypred=model.predict(X_test)
```

```python
ypred
```

```python
model.predict([[27]])
```

```python
new_ages= pd.DataFrame({'age': [76, 34, 9, 50]})
new_ages
```

```python
predictions= model.predict(new_ages)
```
